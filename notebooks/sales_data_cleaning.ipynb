```python
import pandas as pd
import numpy as np
import re
```

```python
#load sales data from 2023 and 2024
sales_2023_df = pd.read_csv("../data/raw/veronicas_sales_2023.csv")
sales_2024_df = pd.read_csv("../data/raw/veronicas_sales_2024.csv")
```

```python
display(sales_2023_df)
display(sales_2024_df)
```

```python
# Transpose the datasets to have dates as rows and metrics as columns
# Function to clean and transform sales data
def clean_sales_data(df):
    # Set the first column as the index for transposition
    df = df.set_index('Sales').T
    
    # Reset index and rename it to 'Date'
    df.reset_index(inplace=True)
    df.rename(columns={'index': 'Date'}, inplace=True)
    
    # Convert monetary values to numerical by removing '$' and commas, then converting to float
    for col in df.columns[1:]:
        df[col] = df[col].replace(r'[\$,]', '', regex=True).astype(float)
    
    return df
```

```python
# Clean both 2023 and 2024 sales datasets
cleaned_sales_2023 = clean_sales_data(sales_2023_df)
cleaned_sales_2024 = clean_sales_data(sales_2024_df)
```

```python
#merge the 2 datasets
merged_sales_df = pd.concat([cleaned_sales_2023, cleaned_sales_2024], ignore_index=True)
```

```python
# Function to convert column names to snake case and avoid double underscores
def to_snake_case(column_name):
    column_name = re.sub(r'([A-Z])', r'_\1', column_name)    # Insert underscores before uppercase letters
    column_name = re.sub(r'\s+', '_', column_name)           # Replace spaces with underscores
    column_name = re.sub(r'[^a-zA-Z0-9_]', '', column_name)  # Remove special characters
    column_name = re.sub(r'_+', '_', column_name)            # Replace multiple underscores with a single underscore
    column_name = column_name.lower().strip('_')             # Convert to lowercase and strip leading/trailing underscores
    return column_name
```

```python
# Apply the function to all column names
merged_sales_df.columns = [to_snake_case(col) for col in merged_sales_df.columns]

# Display updated column names
merged_sales_df.columns
```

```python
#function to format date column to '%Y-%m-%d'
def format_date_column(df, column_name, date_format='%Y-%m-%d'):
    """
    Converts a specified date column in a DataFrame to a given date format.
    
    Parameters:
    - df (pd.DataFrame): The DataFrame containing the date column.
    - column_name (str): The name of the date column to format.
    - date_format (str): The desired date format (default is '%Y-%m-%d').
    
    Returns:
    - pd.DataFrame: The DataFrame with the formatted date column.
    """
    # Ensure the column exists
    if column_name not in df.columns:
        raise ValueError(f"Column '{column_name}' not found in DataFrame.")
    
    # Convert the column to datetime format (if not already)
    df[column_name] = pd.to_datetime(df[column_name], errors='coerce')
    
    # Apply the specified date format using strftime
    df[column_name] = df[column_name].dt.strftime(date_format)
    
    return df
```

```python
#applying the date formatting
merged_sales_df = format_date_column(merged_sales_df, 'date', '%Y-%m-%d')
```
